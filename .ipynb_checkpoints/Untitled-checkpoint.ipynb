```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

from twistpy.convenience import generate_multicomponent_synthetics
```

First, we define the coordinates of the source and the receivers:

```python
source_coordinates = np.asarray([0., 0., 0.])

n_r = 50  # Number of receivers
rx = 50 * np.ones((n_r, 1))  # X coordinates of receivers
ry = 50 * np.ones((n_r, 1))  # Y coordinates of receivers
rz = np.reshape(np.linspace(-24, 24, n_r), (n_r, 1))

receiver_coordinates = np.concatenate((rx, ry, rz), axis=1)
```

Additionally, we need to specify the center frequency of the wavelet that we want to model (will be a Ricker wavelet), the velocity of the wave that we are modelling, and the simulation time and sampling rate:

```python
velocity_P = 6000  # P-wave velocity in m/s
velocity_SV = 3000 # SV-wave velocity in m/s
velocity_SH = 2800 # SH-wave velocity in m/s
wavelet_center_frequency = 200  # Wavelet center frequency in Hz

dt = 2.5e-4  # Sampling interval of the synthetic seismograms (in s)
tmax = 0.08  # Maximum time to simulate (in s)

t = np.arange(0, tmax, dt)
```

Now, we can generate the synthetics for P, SV, and SH waves separately:

```python
s_P = generate_multicomponent_synthetics(source_coordinates, 
                                         receiver_coordinates, 
                                         t, 
                                         velocity_P, 
                                         wavelet_center_frequency, 
                                         wave_type='P') # P-wave seismograms
s_SV = generate_multicomponent_synthetics(source_coordinates, 
                                         receiver_coordinates, 
                                         t, 
                                         velocity_SV, 
                                         wavelet_center_frequency, 
                                         wave_type='SV') # SV-wave seismograms
s_SH = generate_multicomponent_synthetics(source_coordinates, 
                                         receiver_coordinates, 
                                         t, 
                                         velocity_SH, 
                                         wavelet_center_frequency, 
                                         wave_type='SH') # SH-wave seismogram
```

Now we plot the seismograms:

```python
fig1, ax1 = plt.subplots(1, 3)
components = ['X', 'Y', 'Z']
clim = [-0.4, 0.4]
plt.suptitle('P-wave')

for n, ax in enumerate(ax1):
    ax.imshow(s_P[n].T,
              extent=[1, n_r, t[0], t[-1]],
              cmap='seismic',
              aspect='auto',
              interpolation='None',
              origin='lower',
              vmin=clim[0],
              vmax=clim[1])
    ax.set_xlabel('Receiver number')
    if n == 0:
        ax.set_ylabel('Time (s)')
    ax.set_title(components[n])
    
fig2, ax2 = plt.subplots(1, 3)
clim = [-0.4, 0.4]
plt.suptitle('SV-wave')

for n, ax in enumerate(ax2):
    ax.imshow(s_SV[n].T,
              extent=[1, n_r, t[0], t[-1]],
              cmap='seismic',
              aspect='auto',
              interpolation='None',
              origin='lower',
              vmin=clim[0],
              vmax=clim[1])
    ax.set_xlabel('Receiver number')
    if n == 0:
        ax.set_ylabel('Time (s)')
    ax.set_title(components[n])
    
    
fig3, ax3 = plt.subplots(1, 3)
clim = [-0.4, 0.4]
plt.suptitle('SH-wave')

for n, ax in enumerate(ax3):
    ax.imshow(s_SH[n].T,
              extent=[1, n_r, t[0], t[-1]],
              cmap='seismic',
              aspect='auto',
              interpolation='None',
              origin='lower',
              vmin=clim[0],
              vmax=clim[1])
    ax.set_xlabel('Receiver number')
    if n == 0:
        ax.set_ylabel('Time (s)')
    ax.set_title(components[n])
    
```

We can also superimpose all three wave types to get a combined seismogram containing all wave types:

```python
fig4, ax4 = plt.subplots(1, 3)
clim = [-0.4, 0.4]
plt.suptitle('All waves')

for n, ax in enumerate(ax4):
    ax.imshow(s_P[n].T + s_SV[n].T + s_SH[n].T,
              extent=[1, n_r, t[0], t[-1]],
              cmap='seismic',
              aspect='auto',
              interpolation='None',
              origin='lower',
              vmin=clim[0],
              vmax=clim[1])
    ax.set_xlabel('Receiver number')
    if n == 0:
        ax.set_ylabel('Time (s)')
    ax.set_title(components[n])
```
